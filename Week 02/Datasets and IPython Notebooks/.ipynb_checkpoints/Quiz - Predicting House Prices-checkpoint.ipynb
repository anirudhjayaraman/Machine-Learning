```python
my_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']
```

```python
advanced_features = [
    'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode',
    'condition', # condition of the house
    'grade', # measure of quality of construction
    'waterfront', # waterfront property
    'view', # type of view
    'sqft_above', # sqarefeet above ground
    'sqft_basement', # squarefeet in basement
    'yr_built', # the year built
    'yr_renovated', # the year renovated
    'lat', 'long', # the coordinates of the parcel
    'sqft_living15', # average sqft of 15 nearest neighbors
    'sqft_lot15' # average lot size of 15 nearest neighbors
]
```

```python
train_data, test_data = sales.random_split(0.8, seed = 0)
```

```python
my_features_model = graphlab.linear_regression.create(train_data, target = 'price', features = my_features)
```

```python
advanced_features_model = graphlab.linear_regression.create(train_data, target = 'price', features = advanced_features)
```

```python
print my_features_model.evaluate(test_data)
```

```python
print advanced_features_model.evaluate(test_data)
```

```python
x = my_features_model.evaluate(test_data)['rmse']
print type(x)
```

```python
y = advanced_features_model.evaluate(test_data)['rmse']
```

```python
differenceRMSE = x - y
print differenceRMSE
```
